```python
from pyspark.sql import SparkSession
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import StringIndexer
from pyspark.ml import Pipeline

# Initialize Spark session
spark = SparkSession.builder \
        .appName('FraudDetection') \
        .getOrCreate()
```

```python
from pyspark.ml.feature import VectorAssembler, OneHotEncoder, StringIndexer
```

```python
# Load dataset (replace 'your_dataset.csv' with actual path)
df = spark.read.csv('processed.csv', header=True, inferSchema=True)

# Preview dataset
df.printSchema()
df.show(5)
```

```python
# Index and One-Hot Encode the "type" column
type_indexer = StringIndexer(inputCol='type', outputCol='type_indexed')
type_encoder = OneHotEncoder(inputCol='type_indexed', outputCol='type_encoded')

# Features for the model
feature_columns = ['amount', 'balanceChangeOrg', 'balanceChangeDest', 'type_encoded']

# Assemble features into a single vector
assembler = VectorAssembler(inputCols=feature_columns, outputCol='features')

# Index the target column (isFraud)
label_indexer = StringIndexer(inputCol='isFraud', outputCol='label')

# Create the pipeline to streamline the transformations
pipeline = Pipeline(stages=[type_indexer, type_encoder, assembler, label_indexer])

# Apply transformations
df_transformed = pipeline.fit(df).transform(df)
```

```python
# Split data into training and testing sets
train_data, test_data = df_transformed.randomSplit([0.7, 0.3], seed=42)
```

```python
# Initialize Random Forest classifier
rf = RandomForestClassifier(labelCol='label', featuresCol='features', numTrees=100)

# Train the model
rf_model = rf.fit(train_data)
```

```python
# Make predictions on the test set
predictions = rf_model.transform(test_data)

# Evaluate model performance using AUC (ROC)
evaluator = BinaryClassificationEvaluator(labelCol='label', metricName='areaUnderROC')
auc = evaluator.evaluate(predictions)
print(f"Test AUC: {auc:.4f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

# Convert to Pandas for ROC curve plotting
predictions_pandas = predictions.select('probability', 'label').toPandas()

# Extract probabilities for the positive class
predictions_pandas['fraud_prob'] = predictions_pandas['probability'].apply(lambda x: x[1])

# Plot ROC Curve using Matplotlib
fpr, tpr, _ = roc_curve(predictions_pandas['label'], predictions_pandas['fraud_prob'])
plt.plot(fpr, tpr, label=f'ROC Curve (area = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

```python
# Confusion matrix
predictions.groupBy('label', 'prediction').count().show()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Precision
precision_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="precisionByLabel")
precision = precision_evaluator.evaluate(predictions, {precision_evaluator.metricLabel: 1})  # Precision for the fraud class (1)
print(f"Precision (Fraud): {precision:.4f}")

# F1-Score
f1_evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="f1")
f1 = f1_evaluator.evaluate(predictions)
print(f"F1-Score: {f1:.4f}")
```
